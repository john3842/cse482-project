```python
import nba_py
import nba_py.game
import pandas as pd
```

```python
days = pd.date_range('10/27/2015',periods=171,frequency='D')
days
```

```python
games = []
for date in days:
    print(date)
    scores = nba_py.Scoreboard(month=date.month, day=date.day, year=date.year, league_id='00', offset=0)

    for game in scores.available()['GAME_ID']:
        stats = (nba_py.game.Boxscore(game,season='2015-16',season_type='Regular Season').team_stats())
        advanced_stats = (nba_py.game.BoxscoreAdvanced(game,season='2015-16',season_type='Regular Season').sql_team_advanced())
        stats['DATE'] = date
        stats['OFFICIAL_GAME_ID'] = stats['GAME_ID']
        stats['GAME_ID'] = pd.to_numeric(stats['GAME_ID']) - 21500001
        for col in advanced_stats:
            if col not in stats:
                stats[col] = advanced_stats[col]
        
        games.append(stats)
```

```python
days = pd.date_range('10/25/2016',periods=170,frequency='D')
days
```

```python
for date in days:
    print(date)
    scores = nba_py.Scoreboard(month=date.month, day=date.day, year=date.year, league_id='00', offset=0)

    for game in scores.available()['GAME_ID']:
        stats = (nba_py.game.Boxscore(game,season='2016-17',season_type='Regular Season').team_stats())
        advanced_stats = (nba_py.game.BoxscoreAdvanced(game,season='2016-17',season_type='Regular Season').sql_team_advanced())
        stats['DATE'] = date
        stats['OFFICIAL_GAME_ID'] = stats['GAME_ID']
        stats['GAME_ID'] = pd.to_numeric(stats['GAME_ID']) - 21600001 + 1230
        for col in advanced_stats:
            if col not in stats:
                stats[col] = advanced_stats[col]
        
        games.append(stats)

print("done collecting")
```

```python
filepath = 'all_games_data.csv'
test = pd.concat(games)

print("done concat")
test.index = test['GAME_ID']

print("done formatting")
test.sort_index(axis=0,inplace=True)

test.drop(test.tail(8).index,inplace=True)

print("done sorting")
test.to_csv(filepath)

print("DONE")
```
