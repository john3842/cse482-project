```python
import pandas as pd

all_games = pd.read_csv('all_games_data.csv')

all_games.head()
```

```python
all_games['TEAM_CITY'] = all_games['TEAM_CITY'].replace(['LA'], 'Los Angeles')
```

```python
all_games['TEAM'] = all_games['TEAM_CITY']+' '+all_games['TEAM_NAME']
all_games['DATE'] = pd.to_datetime(all_games['DATE'])
```

```python
ha = pd.read_csv('home_away.csv')

indices = [i for i in range(0,2460)]

ha['Date'] = pd.to_datetime(ha['Date'])
ha = ha.sort_values(['Date'], ascending = [True])
ha = ha.set_index([indices])

ha.to_csv('home_away.csv',index=False)

home = []
for row in all_games.iterrows():
    x = ha.loc[ha['Date'] == row[1]['DATE']]
    y = list(x['Visitor/Neutral'])
    if row[1]['TEAM'] in y:
        home.append(0)
    else:
        home.append(1)

all_games['HOME'] = home

all_games[0:10]
```

```python
import numpy as np

teams_dict = {key: value for key, value in all_games.groupby('TEAM_ABBREVIATION')}

teams_dict['LAC'][0:5]
```

```python
indices = [i for i in range(0,164)]

for key in teams_dict:
    teams_dict[key] = teams_dict[key].set_index([indices])
    
teams_dict['GSW'][0:21]
```

```python
#cols = [col for col in all_games.columns if col not in ['GAME_ID', 'OFFICIAL_GAME_ID','TEAM_ID','TEAM_NAME','TEAM_ABBREVIATION','TEAM_CITY','MIN','DATE']]
columns = []
for col in all_games.columns:
    if col not in ['GAME_ID', 'OFFICIAL_GAME_ID','TEAM_ID','TEAM','TEAM_NAME','TEAM_ABBREVIATION','TEAM_CITY','MIN','DATE']:
        columns.append(col+'_0')
        columns.append(col+'_1')
        
columns.append('WIN_PCT_0')
columns.append('WIN_PCT_1')
columns.append('WINNER')
```

```python
teams = {}
count = 0
num = 15

for row in all_games.iterrows():
    if row[1]['TEAM_ABBREVIATION'] in teams:
        teams[row[1]['TEAM_ABBREVIATION']] += 1
    else:
        teams[row[1]['TEAM_ABBREVIATION']] = 1
        
    if all(value >= num for value in teams.values()):
        break
    count+=1
    
count
```

```python
games = []
game = pd.DataFrame(columns = columns)

zero_one = 0
winner = -1
for row in all_games[count+2:].iterrows():    
    idx = teams_dict[row[1]['TEAM_ABBREVIATION']].index[teams_dict[row[1]['TEAM_ABBREVIATION']]['GAME_ID'] == row[1]['GAME_ID']]
    idx = idx.tolist()[0]
    last_ten = teams_dict[row[1]['TEAM_ABBREVIATION']][idx-num:idx]
    current_game = teams_dict[row[1]['TEAM_ABBREVIATION']][idx:idx+1]

    if zero_one == 0:
        #print(row[1]['GAME_ID'])
        #print(row[1]['DATE'])
        #print(row[1]['TEAM_ABBREVIATION'])
        
        for i in range(0,len(columns),2):
            if columns[i] not in ['WINNER','WIN_PCT_0']:
                game.at[0,columns[i]] = last_ten[columns[i][:-2]].mean()
                
        game.at[0,'HOME_0'] = current_game['HOME'].values[0]
        game.at[0,'WIN_PCT_0'] = len(last_ten[last_ten['PLUS_MINUS'] > 0]) / len(last_ten)
        
        winner = row[1]['PLUS_MINUS']
        #print(winner)
        if winner > 0:
            game.at[0,'WINNER'] = 0
        else:
            game.at[0,'WINNER'] = 1
        
        zero_one = 1
        
    else:
        #print(row[1]['GAME_ID'])
        #print(row[1]['TEAM_ABBREVIATION'])
        for i in range(1,len(columns),2):
            if columns[i] not in ['WINNER','WIN_PCT_1']:
                game.at[0,columns[i]] = last_ten[columns[i][:-2]].mean()
    
        game.at[0,'HOME_1'] = current_game['HOME'].values[0]
        game.at[0,'WIN_PCT_1'] = len(last_ten[last_ten['PLUS_MINUS']> 0]) / len(last_ten)
        
        games.append(game)
        game = pd.DataFrame(columns = columns)
        zero_one = 0
        winner = -1
```

```python
indices = [i for i in range(0,len(games))]
feature_vector = pd.concat(games)
feature_vector = feature_vector.set_index([indices])

filepath = 'feature_vectors.csv'
feature_vector.to_csv(filepath)

feature_vector.head()
```
